```python
ws_awsl, wg_awsl, wd_awsl, years_awsl, months_awsl = get_stats(True)
```

```python
def plot_bar_tuple(feature):
    X = [f'[{rng[0]}, {rng[1]}[' for rng in feature.keys()]
    X = X[:-1] + ['missing']
    plt.bar(X, feature.values())
    if (0,15) in X:
        plt.xticks(X[::3], rotation = 45)
    else:
        plt.xticks(X, rotation=45)
    plt.tight_layout()
    plt.yscale('log')
```

```python
def plot_bar_int(feature):
    plt.bar(feature.keys(), feature.values())
    if any(key <= 12 for key in feature.keys()):
        X = [calendar.month_name[num] for num in feature.keys()]
        plt.xticks(list(feature.keys()), X, rotation = 45)
    else:
        plt.xticks(rotation = 45)
    plt.tight_layout()
    plt.yscale('log')
```

```python
plot_bar_tuple(ws)
plot_bar_tuple(ws_awsl)
plt.show()
```

```python
plot_bar_tuple(wg)
plot_bar_tuple(wg_awsl)
plt.show()
```

```python
plot_bar_tuple(wd)
plot_bar_tuple(wd_awsl)
plt.show()
```

```python
plot_bar_int(years)
plot_bar_int(years_awsl)
plt.show()
```

```python
plot_bar_int(months)
plot_bar_int(months_awsl)
plt.show()
```
