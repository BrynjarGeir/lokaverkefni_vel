```python
import pandas as pd, rasterio
from utils.elevation import generateLandscapeDistrbution, generateElevationDistribution

mergedPath = "E:/Skóli/HÍ/Vélaverkfræði Master Hí/Lokaverkefni/Data/merged-full-25ms-24hr-28-2-24.feather"
elevationPath = "E:/Skóli/HÍ/Vélaverkfræði Master Hí/Lokaverkefni/Data/IslandsDEMv1.0_20x20m_isn93_zmasl.tif"
outputPath = "E:/Skóli/HÍ/Vélaverkfræði Master Hí/Lokaverkefni/Data/merged-w-landscape-full-25ms-24hr-28-2-24.feather"
```

```python
df = pd.read_feather(mergedPath)

with rasterio.open(elevationPath) as dataset:
    elevation = dataset.read(1)
    index = dataset.index
    transform = dataset.transform

df['landscape_points'] = df.apply(generateLandscapeDistrbution, axis = 1)

df['elevations']  = df.apply(generateElevationDistribution, args = (transform, index, elevation), axis = 1)
```
